## 3. решить систему:

$
\begin{cases}
x^2 + yx - 9 = 0\\ 
x - y/5 = 0\\
\end{cases}
$

#### решение в Python:

```python
print('ax^2 + bx + c = 0:')
a = 6
b = 0
c = -9

def systems_of_equations(a, b, c, y=False):
    discr = b ** 2 - 4 * a * c
    print('Дискриминант D = %.2f' % discr)
    if discr > 0:
        x1 = (-b + np.sqrt(discr)) / (2 * a)
        x2 = (-b - np.sqrt(discr)) / (2 * a)
        print(f'x1 = {x1:0.3f} \nx2 = {x2:0.3f}')
    elif discr == 0:
        x = -b / (2 * a)
        print(f'x ={x:0.3f}')
    else:
        print('Корней нет')
    if y:
        print(f'y1 = {5*x1:0.3f}\ny2 = {5*x2:0.3f}')
        
systems_of_equations(a, b, c, y=True)
```

#### Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
%matplotlib inline
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12, 5))

x = np.linspace(-5, 5, 521)
plt.plot(x, np.cos(x * 1), color='green', label='cos(x$\cdot$1)')
plt.plot(x, np.cos(x * 2), color='red', label='cos(x$\cdot$2)')
plt.title('две кривые y(x)', size=18)
plt.legend(loc='lower center', fontsize = 14);
```
