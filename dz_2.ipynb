## 1. Напишите уравнение параболы, проходящей через три точки (x,y):
#### (1,2), (3,10), (5,1)

уравнение параболы:

$y = ax^2 + bx + c$

подставляем наши значения и получаем систему:

$
\begin{cases}
a + b + c = 2\\ 
9a + 3b + c = 10\\
25a + 5b + c = 1
\end{cases}
$

ну и решаем теперь

```python
A, B = np.array([[1, 1, 1,], [9, 3, 1], [25, 5, 1]]), np.array([2, 10, 1])
np.round(np.linalg.solve(A, B), 2)
```

построим график, полюбуемся

```python
plt.style.use('classic')
plt.figure(figsize=(12, 5))
x = np.linspace(0, 6, 1001)
plt.plot(x, -2.12*x**2 + 12.5*x - 8.38, linestyle = ':', linewidth=2, label='парабола наша')
plt.plot(1, 2, marker='*', label='точка(1;2)')
plt.plot(3, 10, marker='*', label='точка(3;10)')
plt.plot(5, 1, marker='*', label='точка(5;1)')
plt.title('график параболы, проходящей через точки (x;y): (1;2), (3;10), (5;1)')
plt.grid()
plt.legend();
```
