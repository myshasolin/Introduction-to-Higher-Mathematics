## 1. Задание (теорема умножения)

**Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.**

$P(A) = \frac{1}{6}$

$P(B) = \frac{1}{6}$

$P(A \cdot B) = P(A) \cdot P(B) = \frac{1}{6} \cdot \frac{1}{6} = \frac{1}{36}$

## 3. Задание

**Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной игральной кости. Обратите внимание на порядок выпадения костей!**

$P(A) = \frac{1}{6}$

$P(B) = \frac{1}{6}$

$P  \Bigl ( P(A \cdot B) + P(B \cdot A) \Bigl ) = \Bigl ( P(A) \cdot P(B) + P(B) \cdot P(A) \Bigl ) = \frac{1}{6} \cdot \frac{1}{6} + \frac{1}{6} \cdot \frac{1}{6} = \frac{1}{36} + \frac{1}{36} = \frac{2}{36} = \frac{1}{18}$

## 4. Задание  (Геометрическая вероятность +интервалы)

**На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии не более 9 см от точки А и не более 15 см от точки В?**

```python
plt.figure(figsize=(10, 3))
AB = np.array([i for i in range(1, 21)])
Ax = np.array([i for i in range(1, 10)])
yB = np.array([i for i in range(1, 16)])
plt.hlines(0.25, AB[0], AB[-1], color='#5D6D7E', label='AB = 20 см')
plt.fill_between(Ax, 1, color='#B8FF33', alpha=0.5, label='Ax = 9 см')
plt.fill_between(5+yB, 0.75, color='#33D7FF', alpha=0.5, label='yB = 15 см')
plt.fill_between([6, 9], 0.5, color='#2222', alpha=0.25)
plt.vlines(6, 0, 0.5, color='red', alpha=0.75)
plt.vlines(9, 0, 0.5, color='red', alpha=0.75)
plt.hlines(0, 9, 6, color='red', alpha=0.75)
plt.hlines(0.5, 9, 6, color='red', alpha=0.75)
plt.scatter(1, 0.25, c='b')
plt.text (1, 0.12, 'A')
plt.scatter(AB[-1], 0.25, c='b')
plt.text (AB[-1], 0.12, 'B')
plt.scatter(Ax[-1], 0.25, c='b')
plt.text (Ax[-1]+0.1, 0.12, 'x')
plt.scatter(yB[5], 0.25, c='b')
plt.text (yB[5]-0.3, 0.12, 'y')
plt.xlim(0, AB[-1]+1)
plt.ylim(-0.1, 1.5)
plt.legend()
plt.yticks([])
plt.title(f'$xy = Ax+ yB-AB = 9+15-20 = 4$\n$P(C) = 4/20 = 2/10 = 1/5 = 0.2$');
```

## 7. Задание**

**Чёрный куб покрасили снаружи белой краской, затем разрезали на 27 одинаковых маленьких кубиков и как попало сложили из них большой куб. С какой вероятностью  все грани этого куба будут белыми?**

Мы получим 4 разных вида маленьких кубиков, кубики эти могут оказаться в любом месте любой своей стороной. Тогда количество возможных вариантов - это количество перестановок внутри группы умноженное на количество возможных поворотов. для каждого вида кубиков оно будет своё:

1) без белых граней у нас остался 1 кубик. Количество вариантов для него = $1! \cdot 24^1$, так как он может вертеться по всякому

2) с одной белой гранью  кубиков 6 шт. Количество вариантов для них = $6! \cdot 4^6$, так как это серединки и нам подойдут 4 варианта

3) с двумя белыми гранями 12 шт. Количество вариантов для них = $12! \cdot 2^{12}$, так как нам подойдёт вариант расположения таких кубиков - или так ли эдак, т.е. всего 2 способа

4) с тремя белыми грянями 8 шт. Количество вариантов для них = $8! \cdot 3^8$, так как это уголки

Все возможные комбинации - это сумма количества перестановок умноженная на все подходящие варианты вращения:

$(1+2+12+8)! \cdot 24^{(1+6+12+8)} = 27! \cdot 24^{27}$

Тогда вероятность собрать белый куб = 
$\frac{1! \cdot 24^1 \cdot 6! \cdot 4^6 \cdot 12! \cdot 2^{12} \cdot 8! \cdot 3^8}{27! \cdot 24^{27}}$

Попробуем посчитать это безумие на Python:

```python
P = (fact(1)*24**1*fact(6)*4**6*fact(12)*2**12*fact(8)*3**8)/fact(27)*24**27
P
```

ну то есть получим мы что-то крайне маловероятное)
